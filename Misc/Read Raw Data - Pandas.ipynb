```python
import requests
import pandas as pd
from pyspark.sql import SparkSession

url = "https://raw.githubusercontent.com/ankitgilhotra/demo_dbx_user_group/main/crop_production.csv"

# Download the file and load into pandas
response = requests.get(url)
response.raise_for_status()

from io import StringIO
csv_data = StringIO(response.text)
pandas_df = pd.read_csv(csv_data)

# Convert to Spark DataFrame for display
spark_df = SparkSession.builder.getOrCreate().createDataFrame(pandas_df)
display(spark_df)
```
