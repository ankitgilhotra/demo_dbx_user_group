```python
from pyspark.sql.functions import col, sum, avg

df = spark.read.table('dbx_demo.silver.cleaned_crop_data')
# display(df.describe())
display(df.select('Crop', 'Area', 'Production', 'State_Name', 'District_Name', 'Crop_Year', 'Season').limit(20))
```

```python
agg_df = df.groupBy('State_Name', 'District_Name', 'Crop_Year', 'Season', 'Crop').agg(
    sum('Area').alias('Total_Area'),
    sum('Production').alias('Total_Production'),
    avg('Production').alias('Avg_Production'),
    avg('Area').alias('Avg_Area')
)
# Save aggregated data as gold_agg table
gold_agg_table = 'dbx_demo.gold_agg.crop_report_agg'
agg_df.write.format('delta').mode('overwrite').saveAsTable(gold_agg_table)
display(agg_df.limit(20))
```
